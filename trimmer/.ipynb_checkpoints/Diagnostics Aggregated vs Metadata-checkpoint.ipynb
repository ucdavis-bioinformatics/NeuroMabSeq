```python
import pandas as pd
```

```python
data = pd.read_csv('../data3/2020-12-09-Aggregated_Sequences.tsv', delimiter='\t')
data.head()
```

```python
data.groupby('Sample_Name').head()
```

```python
len(data['Sample_Name'].unique())
```

```python
show_on_web = data[data['ShowOnWeb']== True].groupby('Sample_Name').count()
show_on_web
```

```python
plate_summary = data[data['ShowOnWeb']== True].groupby(['Sample_Name','plate']).count()

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(plate_summary.reset_index().groupby('plate').count())
```
