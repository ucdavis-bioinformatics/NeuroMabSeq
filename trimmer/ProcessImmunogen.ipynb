```python
# read this file as a csv /Users/keithmitchell/Desktop/Repositories/NeuroMabSeq/trimmer/mAb immunogen sequences proofed 012523 (1).xlsx
df = pd.read_excel('/Users/keithmitchell/Desktop/Repositories/NeuroMabSeq/trimmer/mAb immunogen sequences proofed 012523 (1).xlsx', sheet_name='TrimmerLab_TableCompleted',)
# filter the Target Sequence column > 40 length aand only is letters and not nan
df = df[df['Target Sequence'].str.len() > 100]   
df = df[df['Target Sequence'].str.isalpha()]
df = df[df['Target Sequence'].notna()]
df
```

```python
# print the unique sequences now and aa list of the Neuromabclone in a dictionaary
final_dict = {}
for index, row in df.iterrows():
    if row['Target Sequence'] not in final_dict:
        final_dict[row['Target Sequence']] = [row['NeuroMab Clone - PDF']]
    else:
        final_dict[row['Target Sequence']].append(row['NeuroMab Clone - PDF'])

# pretty print the dictionary
pprint.pprint(final_dict)
```

```python
# iterate through the dictioary aand print a list of the unique sequences and the valaues
key_list = []
val_list = []
for key, value in final_dict.items():
    key_list.append(key)
    val_list.append('+'.join([i.replace("/", "__") for i in value]))
```
