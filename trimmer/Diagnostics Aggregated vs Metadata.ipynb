# Fix the aggregated sequences file based on the metadata

```python
# upload the metadata
metadata = pd.read_csv("../static_data/2021-03-12.tsv", sep='\t')
metadata.head()
```

```python
drop_out_plus_meta = drop_out.merge(metadata, left_on='sample_name', right_on='sample_name')
drop_out_plus_meta.head()
```

```python
drop_out_final_test = drop_out_plus_meta[(drop_out_plus_meta['LCs.Reported']==0) & (drop_out_plus_meta['HCs.Reported']==0)]
drop_out_final_test.head()
```

```python
def plate_to_mab(plate_loc):
    try:
        entry = metadata[metadata['sample_name']==plate_loc]
        return(list(entry['ShowOnWeb'])[0], list(entry['Category'])[0], list(entry['SeqYear'])[0],  list(entry['trimmer_id'])[0])
    except:
        print("ERROR: there is an issue with %s" %(plate_loc))
plate_to_mab('P41_A10')
```

```python
show_on_web_update = []
category_update = []
seqyear_update = []
for index,row in data.iterrows():
    show_on_web_update.append(plate_to_mab(row['Sample_Name'])[0])
    category_update.append(plate_to_mab(row['Sample_Name'])[1])
    seqyear_update.append(plate_to_mab(row['Sample_Name'])[2])
```

```python
data['ShowOnWeb'] = show_on_web_update
data['Category'] = category_update
data['SeqYear'] = seqyear_update
```
